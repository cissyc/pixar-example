# Exploring Pixar data

## Loading data

First load the data in and have a look at dimension and fields available.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# check the files in current directory
import os
os.listdir()
```

```python
# read in pixar data
df_pixar = pd.read_csv("pixar_data.csv")
```

```python
# get dimensions
df_pixar.shape
```

```python
# preview file
df_pixar.head()
```

```python
# overview of fields and data types; note all are numeric data types besides Movie, Domestic % and International %
df_pixar.dtypes
```

```python
df_pixar.describe()
```

Now do some quick data cleaning.

```python
# clean up % columns
df_pixar["Domestic %"] = df_pixar["Domestic %"].str.replace("%", "").astype("float")
df_pixar["International %"] = df_pixar["International %"].str.replace("%", "").astype("float")
```

```python
df_pixar["IMDB Score"] = df_pixar["IMDB Score"] * 10
```

```python
# last entry of file has no revenue data; filter this out
df_pixar.loc[14]
```

```python
# put together another datafram called df_filtered with only entries with all the data
df_filtered = df_pixar.dropna()
df_filtered.shape
```

```python
# set indices
df_pixar.set_index("Movie", inplace = True)
df_filtered.set_index("Movie", inplace = True)
```

## Data visualisation

Plot a few charts to explore our data.

```python
# plot review for each film
# Rotten Tomatoes almost always gives higher score than others
df_pixar_critics_reviews = df_pixar[["RT Score", "IMDB Score", "Metacritic Score"]]
df_pixar_critics_reviews.plot(figsize = (10,6))
```

```python
# Rotten Tomatoes score range generally higher than IMDB score range; Metacritic gives wide range of scores
df_pixar_critics_reviews.plot(kind = "box", figsize = (9,5))
```

```python
# now let's look at revenues, see that revenues are generally on a downward trend
df_filtered_revenue_porportions = df_filtered[["Domestic %", "International %"]]
df_filtered_revenue_porportions.plot(kind = "bar", stacked = True)
```

```python
# compare oscar nominations with actual won oscars
df_filtered_oscars = df_filtered[["Oscars Nominated", "Oscars Won"]]
df_filtered_oscars.plot.bar()
```
